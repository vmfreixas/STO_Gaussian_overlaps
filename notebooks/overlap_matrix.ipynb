```python
import os
os.chdir('/Users/victormanuelfreixaslemus/Desktop/Projects/Git_portafolio/STO_Gaussian_overlaps/STO_Gaussian_overlaps')
from src.molden import read_molden
from src.overlap.build_valence_overlap import build_valence_overlap_matrix, pretty_print_matrix
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Calculate overlap matrix for valence orbitals of benzene
# Rows correspond to STO and columns to GTO

moldenFile = "examples/benzene_cart.mld"
mol = read_molden(moldenFile)

S, rows, cols = build_valence_overlap_matrix(mol, quad_n=128)
pretty_print_matrix(S, rows, cols)
```

```python
np.savetxt("examples/valence_overlap_matrix.txt", S, fmt="%.8f")

# --- Plot ---
plt.figure(figsize=(8, 6))
plt.imshow(S, cmap="coolwarm", interpolation="nearest")
plt.colorbar(label="Overlap value")

plt.title("Valence STOGTO Overlap Matrix")
plt.xticks(range(len(cols)), [c.label for c in cols], rotation=90)
plt.yticks(range(len(rows)), [r.label for r in rows])
plt.tight_layout()

plt.savefig("examples/valence_overlap_matrix.png", dpi=300)
plt.show()
```
